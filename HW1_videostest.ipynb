```python
import pandas as pd
videos = pd.read_csv("https://github.com/fedhere/FDSfE_FBianco/blob/main/data/USyoutubes.csv?raw=true")
videos.shape
videos.describe()
videos.head()

videos.columns
videos[['views', 'likes', 'dislikes', 'comment_count']].corr()

import pylab as pl
pl.imshow(videos[['views', 'likes', 'dislikes', 'comment_count']].corr(),clim=(-1, 1),cmap="RdBu")
pl.xticks([0, 1, 2, 3],['views', 'likes', 'dislikes', 'comment_count'], rotation = 45)
pl.yticks([0, 1, 2, 3],['views', 'likes', 'dislikes', 'comment_count'], rotation = 45)
pl.colorbar();
```
